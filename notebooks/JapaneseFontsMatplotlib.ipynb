```python
import pandas as pd
```

We can produce a simple chart showing some data and labels. If the labels are in English, this is no problem.

```python
planetsEnglish = pd.DataFrame({'mass': [0.330, 4.87 , 5.97],
                   'radius': [2439.7, 6051.8, 6378.1]},
                    index=['Mercury', 'Venus', 'Earth'])
plot = planetsEnglish.plot.pie(y='radius', figsize=(5, 5))
```

But writing the labels in Japanese, you might run into some problems.　The Japanese labels appear as small squares, called 文字化け　(mojibake: garbled text) or sometimes more affectionally 豆腐 (tofu) do to the resemblance to white blocks of soybean protein.

```python
planetsJapanese = pd.DataFrame({'mass': [0.330, 4.87 , 5.97],
                   'radius': [2439.7, 6051.8, 6378.1]},
                    index=['水星', '金星', '地球'])
plot = planetsJapanese.plot.pie(y='radius', figsize=(5, 5))
```

Pandas uses the matplotlib library under the hood, so let's import that and see what fonts we have available. We'll draw each font name to a sample chart along with some sample text.  Many fonts will just show as squares, but the fonts that display correctly can be used on in charts.

```python
import matplotlib
import matplotlib.pyplot as plt

fonts = set([f.name for f in matplotlib.font_manager.fontManager.ttflist])
 
plt.figure(facecolor="white",figsize=(10,len(fonts)/4))
 
for i, font in enumerate(fonts):
    try:
        plt.text(0, i, f"漢字　ひらがな　カタカナ：{font}", fontname=font)
    except:
        print(font)
    
plt.ylim(0, len(fonts))
plt.axis("off")
    
plt.show()
```

```python
plt.figure(facecolor='white',figsize=(6,6))
plt.pie(planetsJapanese['radius'],
        labels=planetsJapanese.index,

        textprops={'fontfamily':'Noto Sans JP','fontsize': 12})
```
