# Typing Japanese Words

If you've ever looked at a Japanese computer keyboard you might have noticed the Japanese symbols in addition to the familiar QWERTY layout.  The truth is, almost no one really uses the Japanese symbols as their input method. Instead, the most common way to input Japanese text is Romaji input (ローマ入力). This involves typing Japanese phonetically using the alphabet and then having software handle the conversion. For example, to type ねこ (neko, cat) you type `neko` with four key presses.  While some Japanese letters can be written with a single keypress in this system, most require two. With the uncommon kana input (かな入力) you would only need to press two keys: `ね こ`

This got me thinking: how much more efficient is kana input compared with romaji input? How many key presses can you save by switching?

```python
import pykakasi
converter = pykakasi.kakasi()

def romajiPressCount(text: str):
    "Take a given text and return how many keypresses needed for romaji input"
    romajiText = convertToRomaji(text)
    return countLetters(romajiText)

def kanaPressCount(text:str):
    hiraganaText = convertToHiragana(text)
    return len(hiraganaText)

def convertToRomaji(text: str):
    results = converter.convert(text)
    return "".join([w['kunrei'] for w in results])

def convertToHiragana(text: str):
    results = converter.convert(text)
    return "".join([w['hira'] for w in results])
    
def countLetters(text):
    "count letters ignoring spaces"
    return len(text.replace(" ", ""))

```

```python
# Tests

assert countLetters("neko ga iru") == 9
assert convertToRomaji("福岡は暑い") == "fukuokahaatui"
assert convertToHiragana("福岡は暑い") == "ふくおかはあつい"
assert romajiPressCount("福岡は暑い") == 13
assert kanaPressCount("福岡は暑い") == 8
```

Using these simple functions, lets compare a few texts and see how much of a difference the two input methods make.

```python
text = "新型コロナの感染の急拡大を受け、岸田総理大臣は22日午後、関係閣僚と対応を協議することにしていて、社会経済活動を維持していくため、濃厚接触者に求める自宅などでの待機期間を原則7日から5日に短縮する方向で調整を進めています"

print(kanaPressCount(text))

print(romajiPressCount(text))
```

Kana input seems like a clear winner. But wait! Our analysis is a little too naive. Not every possible letter is on the kana keyboard layout, there are simple too many to fit. Instead, some of the keys are used together to modify characters. For example, か (ka) becomes が (ga) by pressing the ゛key.
